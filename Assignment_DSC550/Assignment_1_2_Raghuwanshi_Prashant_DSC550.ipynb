####  1) Create an array of size [10,50] and use the random function to fill it with integers between 1 and 500 (inclusive). Load it to a DataFrame.

```python
array_randnums= np.random.randint(1,501, size=(10,50))
array_randnums
```

#### 2) Calculate the sum of each row; the sum of each column; and the sum of all entries.


```python
sumofeachrow = np.sum(array_randnums,axis=1).tolist()
sumofeachrow
```

```python
sumofeachcolumn = np.sum(array_randnums,axis=0).tolist()
sumofeachcolumn
```

#### 3) Determine the minimum; maximum; average; and standard deviation of all entries.

```python
minimum = np.min(array_randnums)
maximum = np.max(array_randnums)
mean = np.mean(array_randnums)
standarddeviation = np.std(array_randnums)
```

```python
minimum, maximum, mean, standarddeviation
```
